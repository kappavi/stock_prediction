First, let us visualze a dataset of s&p stock data from 2013-2018

```python
import kagglehub

# Download latest version
dataset_path = kagglehub.dataset_download("camnugent/sandp500")

print("Path to dataset files:", dataset_path)
```

Now, we want to visualize the data. Let's load the CSV using pandas and then display it using matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(dataset_path + "/all_stocks_5yr.csv")
data.head()


```

```python
# finding the different stocks that are there 
stock_names = set()
for name in data["Name"]:
    stock_names.add(name)
    
print(stock_names)
```

```python
# lets find just AAPL entries 
apple_rows = data.loc[data["Name"] == "AAPL"]
print(apple_rows)
```

```python
# lets plot the data using matplotlib 
x = apple_rows["date"]
open = apple_rows["open"]
close = apple_rows["close"]
plt.plot(open)
plt.plot(close, c = "red")
plt.show()
```
